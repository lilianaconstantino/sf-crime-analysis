```python
# San Francisco Crime Data Analysis

## Objective
#This report analyzes crime data from San Francisco to identify:
#- The most common crime categories
#- Crime distribution by district
#- Crime patterns by time of day

#Data Source: Kaggle SF Crime Classification Dataset
#Tools Used: PostgreSQL, SQL, Python (Pandas, Matplotlib, Seaborn)
```

```python
import pandas as pd
import psycopg2

conn = psycopg2.connect(
    dbname="sfcrime",
    user="lilianaconstantino",
    host="localhost"
)
```

```python
query = """
SELECT *
FROM train
LIMIT 10;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT COUNT(*) AS total_records
FROM train;
"""

pd.read_sql(query, conn)
```

```python
# Top Crime Categories 
```

```python
query = """
SELECT category, COUNT(*) AS crime_count
FROM train
GROUP BY category
ORDER BY crime_count DESC;
"""

top_crimes = pd.read_sql(query, conn)
top_crimes
```

```python
top_crimes.to_csv("top_crimes.csv", index=False)
```

```python
# Crime distribution by Police District 
```

```python
query = """
SELECT pd_district, COUNT(*) AS crime_count
FROM train
GROUP BY pd_district
ORDER BY crime_count DESC;
"""

district_crime = pd.read_sql(query, conn)
district_crime
```

```python
district_crime.to_csv("crime_by_district.csv", index=False)
```

```python
# Crime by Day of Week
```

```python
query = """
SELECT dayofweek, COUNT(*) AS crime_count
FROM train
GROUP BY dayofweek
ORDER BY crime_count DESC;
"""

crime_by_day_of_week = pd.read_sql(query, conn)
crime_by_day_of_week
```

```python
crime_by_day_of_week.to_csv("crime_by_day_of_week.csv", index=False)
```

```python
# Crime by Hour of Day 
```

```python
query = """
SELECT EXTRACT(HOUR FROM dates) AS hour, COUNT(*) AS crime_count
FROM train
GROUP BY hour
ORDER BY hour;
"""


crime_by_hour_of_day = pd.read_sql(query, conn)
crime_by_hour_of_day
```

```python
crime_by_hour_of_day.to_csv("crime_by_hour_of_day.csv", index=False)
```

```python
# Day + Hour + Crime count
```

```python
query = """
SELECT 
    dayofweek,
    EXTRACT(HOUR FROM dates) AS hour,
    COUNT(*) AS crime_count
FROM train
GROUP BY dayofweek, hour
ORDER BY dayofweek, hour;
"""

df = pd.read_sql(query, conn)

df.head()
```

```python
df.to_csv("crime_by_day_hour.csv", index=False)
```
