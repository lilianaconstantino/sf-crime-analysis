```python
# San Francisco Crime Data Analysis

## Objective
#This report analyzes crime data from San Francisco to identify:
#- The most common crime categories
#- Crime distribution by district
#- Crime patterns by time of day

#Data Source: Kaggle SF Crime Classification Dataset
#Tools Used: PostgreSQL, SQL, Python (Pandas, Matplotlib, Seaborn)
```

```python
import pandas as pd
import psycopg2

conn = psycopg2.connect(
    dbname="sfcrime",
    user="lilianaconstantino",
    host="localhost"
)
```

```python
query = """
SELECT column_name, data_type
FROM information_schema.columns
WHERE table_name = 'train';
"""

pd.read_sql(query, conn)
```

```python
#The dataset contains 878,049 crime records.
```

```python
query = """
SELECT category, COUNT(*) AS total
FROM train
GROUP BY category
ORDER BY total DESC
LIMIT 10;
"""

top_crimes = pd.read_sql(query, conn)
top_crimes
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.bar(top_crimes['category'], top_crimes['total'])
plt.xticks(rotation=45)
plt.title("Top 10 Crime Categories")
plt.show()
```

```python
#Larceny/Theft is the most common crime category, significantly exceeding other crime types.
```
